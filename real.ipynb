```python
# 16 . remove_duplicates
def remove_duplicates(A):
    ###given a vector of class instances, it returns another vector with the
    ###duplicates removed

    B = remove_dup_pos(1, A)
 
    ###여기서 remove_dup_pos 의 이름이랑 인풋 아웃풋 확인
    return B

def remove_single(ele, A):

    ###removes all occurrences of element ele in A
    B = [];
    for k in range(len(A)):
        if not(ele == A[k]):
            B.append(A[k])
    return B

def remove_dup_pos(pos, A):
###removes all duplicates of A from position pos downwards

    if(pos >= len(A)):
        B = A;
    else:
        ###removes duplicates of element in position pos and call the function
        ###again
        newA = A[0:(pos+1)]
        newA.append( remove_single(A[pos+1], A[(pos+2):(len(A)-1)]))
        B = remove_dup_pos(pos+1, newA);

    return B
```

```python
# Generated with SMOP  0.41
# from libsmop import *
# 19. tensor.m

def tensor(*args):
    nargin = len(args)

    #computes the tensor product of the input
    A=1
# tensor.m:3
    for k in np.arange((1,nargin)).reshape(-1):
        A=np.kron(A,args[k])
# tensor.m:5
    
    return A
    
if __name__ == '__main__':
    pass
    
```

```python
# 8. generalized 3 CHSH

def generalized_3CHSH_test():
# def allNumbers(num=None,d=None,*args,**kwargs):
# generates the Bell functional (in notation B(x,b,z)) for the conditioned
# 3CHSH inequality disp((dubbed $\mathscr{T}(P) in the article)

# first, generate all correlators

    E = np.empty((3,6),dtype='object')

    for x in range(3):
        for z in range(6):
            E[x,z]=(2*ket(x+2,4)-ket(1,4))*(2*bra(z+2,7)-bra(1,7))

    # next, generate 3CHSHs
    B=np.zeros((4,4,7))
    for b in range(4):
        # extract two bits from b
        b1=cifra(b,1,2)
        b2=cifra(b,2,2)
        B[:,b,:]=(-1)**b2*E[1-1,1-1]+(-1)**b2*E[1-1,2-1]+(-1)**(b1)*E[2-1,1-1]-(-1)**(b1)*E[2-1,2-1]+\
            (-1)**b2*E[1-1,3-1]+(-1)**b2*E[1-1,4-1]-(-1)**(b1+b2)*E[3-1,3-1]+(-1)**(b1+b2)*E[3-1,4-1]+\
            (-1)**(b1)*E[2-1,5-1]+(-1)**(b1)*E[2-1,6-1]-(-1)**(b1+b2)*E[3-1,5-1]+(-1)**(b1+b2)*E[3-1,6-1]
    return B
```

```python
# 17 . Seq
class Seq:
    ###class for operator monomials of non-commutative projector variables
    
    def __init__(self, ops):
         ###%list of operators. The associated monomial is
         ####%ops(1)ops(2)...ops(length(ops))
         self.ops = ops
    

    def ope_multiply(self,seq, ope):  
        ###%multiply sequence seq by operator ope on the right
        ops = seq.ops
        L = len(ops)
        
        if(ops[L-1].meas and ope.meas):
            if(ops[L-1].setting == ope.setting):
                ###if the settings of the last operator of the list and the new 
                ###operator are the same, the sequence doesn't change
                seq2 = seq
            else:
                ###%if they are different, add new operator to the
                ###%sequence list
                ops.append(ope)
                seq2= Seq(ops)
        elif(ops[L-1].id):
            ###%if the last element equals the identity, then L=1 and
            ###%seq2=ope
            empty_ops = []
            empty_ops.append(ope)
            seq2 = Seq(empty_ops)
        else:
            ####%if not, then ope is the identity, so the sequence stays
            ###%the same
            seq2 = seq
            
  
        return seq2
        
        
        
    def fmtimes(self,seq1, seq2):
                ###product of two operator monomials
                
        ops2 = seq2.ops
        final = seq1
        for k in range(len(ops2)):
            ope = ops2[k]
            final = self.ope_multiply(final, ope)
                    
        return final
        
    def adjoint(self,seq):
        ####%adjoint of a monomial
        ops = seq.ops
        ####%invert sequence
        ops = ops.reverse()
        ### %return new sequence
        seq_ad = Seq(ops)
    
        return seq_ad


    def disp(self,seq):
       ##%display a monomial
       cadena = ''
       for k in range(len(seq.ops)):
           cadena = cadena+ str(seq.ops[k])

       print(cadena)
       
       return 
        
    def eq(self,seq1, seq2):
        ###%determine if two monomials are the same            
        if (len(seq1.ops)==len(seq2.ops)):
            bol = True
            for k in range(len(seq1.ops)):
                bol = bol and (seq1.ops[k] == seq2.ops[k])
        else:
            bol = False
                
        return int(bol) 

    def id():
        # generates the identity monomial
        return Seq(Ope(0))

    def meas(setting):
        # generates the monomial corresponding to a single projection
        # operator E_setting
        return Seq([Ope(setting)])
        

```

```python
# Generated with SMOP  0.41
# from libsmop import *
# 1. all_seq.m


def all_seq(n=None,num_settings=None,*args,**kwargs):
    

    #generates a vector with all operator monomials of length n or lower of the
#projector variables E_1,...,E_num_settings
    
    list_seq1=np.concatenate([a for a in [Seq.id]])
# all_seq.m:5
    for setting in np.arange(1,num_settings).reshape(-1):
        list_seq1=np.concatenate([a for a in [list_seq1,Seq.meas(setting)]])
# all_seq.m:7
    
    if n == 1:
        list_seq=np.asanyarray(list_seq1).copy(order="F")
# all_seq.m:12
    else:
        list_minus_1=all_seq(n - 1,num_settings)
# all_seq.m:14
        list_seq=[]
# all_seq.m:15
        for ope1 in list_seq1.reshape(-1):
            for ope2 in list_minus_1.reshape(-1):
                ope12=np.dot(ope1,ope2)
# all_seq.m:18
                list_seq=np.concatenate([a for a in[list_seq,ope12]])
# all_seq.m:19
        #remove duplicate operators
        list_seq=remove_duplicates(list_seq)
# all_seq.m:24
    
    return list_seq
    
```

```python
# Generated with SMOP  0.41
# from libsmop import *
# 6. gen_mat.m

def adjugate(m):
    c = np.linalg.inv(m).T * np.linalg.det(m)
    return c.T

def gen_mat(seq=None,n=None,num_settings=None,*args,**kwargs):

    #given an operator monomial seq, it generates a matrix M, with rows and 
#columns labeled by the monomials of degree n or lower of a unital 
#*-algebra with projection generators E_1,...,E_{num_settings}, such that 
#M_{a,a'}=\delta_{a'a^\dagger, seq}
    
    #generate a list of all operator monomials of degree n or lower
    list_seq=all_seq(n,num_settings)
# gen_mat.m:8
    L=len(list_seq)
# gen_mat.m:9
    M=np.zeros(L,L)
# gen_mat.m:10
    for k1 in np.arange(1,L).reshape(-1):
        for k2 in np.arange(1,L).reshape(-1):
            ope1=list_seq(k1)
# gen_mat.m:13
            ope2=list_seq(k2)
# gen_mat.m:14
            if (seq == (np.dot(ope1,adjugate(ope2)))):
                M[k2,k1]=1
# gen_mat.m:16
    
    
    
    return M
    
```
